```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import make_circles
```

```python
from sklearn.datasets import make_moons
moons = make_moons(n_samples = 1000,                   noise=.1,                   random_state=42)[0]
plt.scatter(moons[:,0], moons[:,1]);
```

# DBSCAN

```python
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
moons = scaler.fit_transform(moons)
plt.scatter(moons[:,0], moons[:,1]);
```

```python
min_samples = 30
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors.fit(moons)
distances, indices = neighbors.kneighbors(moons)
distances[:5]
```

```python
sorted_distances = np.sort(distances[:, min_samples-1])
```

```python
plt.plot(sorted_distances)
plt.grid();
```

```python
# Instantiate & fit clustering - this is done in one step for DBSCAN
dbs = DBSCAN(eps = 0.36, min_samples = min_samples).fit(moons)

# I tinkered with this by trial and error to get it perfect!
```

```python
plt.scatter(moons[:,0], moons[:,1], c=dbs.labels_);
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# Visualize Dendogram
plt.figure(figsize = (15, 5))
sch.dendrogram(sch.linkage(moons, method = 'ward'))
plt.xlabel('Data Points');
```

```python
# Instantiate clustering
hc = AgglomerativeClustering(n_clusters = 2)
hc.fit(moons)
```

```python
plt.scatter(moons[:,0], moons[:,1], c=hc.labels_);
```
