```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path ="https://docs.google.com/spreadsheets/d/e/2PACX-1vR4AOPKKW6zLpiLygapfcRy1vJjV130vQGRCy2v4Tx8vJARHy9K28TWVFboFOgxiPAGEu5R6nWZw3Qw/pub?output=xlsx"
cereal = pd.read_excel(path, header=1)
cereal.head()
```

```python
cereal.info()
```

# Scatter Plot in Python

```python
cereal.plot.scatter(x = 'grams of sugars', y = 'calories per serving');
```

This is a positive correlation.

On average, cereals with more sugar also have more calories and cereals with more calories also have more sugar.

# Using a Groupby to make a bar chart

```python
manu_cal = cereal.groupby('Manufacturer')['calories per serving'].mean()
manu_cal
```

# Vertical Bar Chart

```python
plt.bar(manu_cal.index, manu_cal.values)
plt.ylabel('Average Calories')
plt.xlabel('Manufacturer')
plt.xticks(rotation=45);
```

# Horizontal Bar Chart

```python
plt.barh(manu_cal.index, manu_cal.values)
plt.ylabel('Manufacturer')
plt.xlabel('Average Calories');
```

# Sorted Horizontal Bar Chart

```python
manu_cal2 = manu_cal.sort_values()
plt.barh(manu_cal2.index, manu_cal2.values)
plt.ylabel('Manufacturer')
plt.xlabel('Average Calories');
```

Nabisco has the least amount of calories per serving, while Ralston Purina has the most calories per serving.

## Seaborn allows you to create more advanced plots and trivariate analysis!

```python
import seaborn as sns
```

```python
manufact = cereal.groupby(['Manufacturer', 'type']).mean().reset_index()
manufact
```

```python
manufact = manufact.sort_values(by = 'calories per serving')
manufact
```

# Comparing 3 Variables in a Bar Plot

```python
sns.barplot(data = manufact, x = 'calories per serving', y = 'Manufacturer', hue = 'type')
plt.title('Average Calories in Cereal by Manufacturer and Type')
plt.legend(bbox_to_anchor=(1,1));
```

Only Nabisco, Quaker Oats and American Home Food Products makes hot cereals. Hot cereals have around 100 calories per serving. Nabisco cold cereals have the least amount of calories per serving.

# Color Coding Categories in a Scatterplot

```python
sns.scatterplot(data = cereal, x = 'calories per serving', y = 'grams of sugars', hue = 'type')
plt.legend();
```

This shows a positive correlation. Hot cereals have around 100 calories per serving, while cold cereals range from 50 to 160 calories per serving. Also, most of the cold cereals have more grams of sugar.

# Creating a Pairplot Using Seaborn

```python
sns.pairplot(cereal.iloc[:, [3, 7, 8, 9, 10]])
```

Here you can see the correlations and distributions together in one plot.
