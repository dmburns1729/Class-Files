## Conditionals <a name="conditionals">

Python has three conditional statements:
- If Statements
    - executes only if the condition is true.

- Else Statements
    - used when you have two different code blocks to run
    - if the condition is True, the code inside the "if" block is executed.
    - if the condition is False, the code in the 'else' block is executed instead.

- Elif Statements
    - used when you have multiple conditions which can be True or False
    - the "if" condition is tested first. If true, python will execute the code under the if block.
    - if the first condition is False, it will check for an "elif" condition.
    - if that is also False, then the "else" block of code will be executed.
    - you can have multiple "elif" conditions.

```python
# If Statement

a = 0.0065
b = 0.0257

if a < b:
  print('a is less than b')
```

```python
# Else Statement

if a > b:
  print('a is greater than b ')
else:
  print('a is not greater than b')
```

```python
# Elif Statement

if a == b:
  print('a is equal to b')
elif a > b:
  print('a is greater than b')
else:
  print('a is less than b')
```

Combining Conditionals

- and (&): if both conditions must be true
- or (|): if either condition must be true

```python
num = 6
if (num > 6) & (num % 2 != 0):
  print('number is greater than to 6 and odd')
elif (num >= 6):
  print('number is greater than or equal to 6 and even')
elif (num % 2 != 0):
  print('number is less than 6 and odd')
else:
  print('number is less than 6 and even')
```

```python
num = 11
if (num >27) | (num % 2 == 0):
  print('number is either greater than 27, even, or both')
```

## Loops <a name="loops">

There are two kinds of loops in Python
- While Loop: while the condition is True, repeat the operation
- For Loop: for each object in the iterable, repeat the operation


- **Iterable**:
    - a variable type that holds multiple entries.
        - Any variable you can loop through is an iterable:
            - Lists
            - Dictionaries
            - Tuples



```python
# while loop
i = 11

while i > 5:
  i = i - 1
  print(i)
```

```python
# for loop through list
for item in list_var1:
  print(item)
```

```python
# for loop through string
for letter in sherlin_str:
  print(letter)
```

```python
# for loop through dict keys
for key in dict_var.keys():
  print(key)
```

```python
# for loop through dict values
for d in dict_var:
  print(dict_var[d])
```

```python
# for loop through keys and values
for d in dict_var.items():
  print(d)
```

```python
# for loop through keys AND values.
for key, value in dict_var.items():
  print(f'{key} = {value}')
```
