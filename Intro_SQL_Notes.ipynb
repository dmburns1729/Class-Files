Aggregates
If we want to get aggregate statistics on numeric columns, we can use keywords like:

AVG: Average

SUM: Sum of values

MIN: Minimum value

MAX: Maximum value

COUNT: Number of values

DISTINCT: Like value_counts



SELECT AVG(column)

FROM table;

Specify table


SELECT person.name

FROM person;

Specify table as alias

SELECT p.name

FROM person AS p;

Can skip the as

SELECT p.name

FROM person p;



```python
# Use groupby function

sql = """SELECT LocatedInRuralOrUrban, AVG(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY LocatedInRuralOrUrban;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT PostOfficeName, MAX(`DistrictHQDistance (in Km)`)
FROM school_details
GROUP BY PostOfficeName;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
ORDER BY COUNT(*) DESC;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT CRCName, COUNT(*) AS NumberofSchools
FROM school_details
GROUP BY CRCName
ORDER BY NumberofSchools DESC;
"""
pd.read_sql_query(sql, engine)
```

```python
# Having is like WHERE but for a GROUP BY

sql = """SELECT CRCName, COUNT(*)
FROM school_details
GROUP BY CRCName
HAVING COUNT(*) > 20;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT CRCName, COUNT(*) AS NumberofSchools
FROM school_details
GROUP BY CRCName
HAVING NumberofSchools > 20;
"""
pd.read_sql_query(sql, engine)
```

```python
sql = """SELECT SubdivisionName, AVG(`DISTRICTHQDistance (in km)`) AS AverageDistance
FROM school_details
GROUP BY SubdivisionName
HAVING AverageDistance > 30;
"""
pd.read_sql_query(sql, engine)
```

```python


SELECT *
FROM Person p INNER JOIN Location l
  ON l.id = p.birth_location_id
```

```python
SELECT p.name, l.city
FROM Person p LEFT JOIN Location l
  ON l.id = p.birth_location_id
```

```python
SELECT MAX(p.height), l.city
FROM Person p
INNER JOIN Location l ON l.id = p.birth_location_id
GROUP BY l.city
HAVING MAX(p.height) < 65 ORDER BY l.city
```

```python
# Full join or full outer join don't work in MySQL.  This is a work around

SELECT teacher.LastName, subject.SubjectName
FROM teacher
LEFT JOIN subject
ON teacher.SubjectID = subject.SubjectID
UNION
SELECT teacher.LastName, subject.SubjectName
FROM teacher
RIGHT JOIN subject ON teacher.SubjectID = subject.SubjectID;
```

Subquery in a FROM Clause

Next, lets say we want to know the average value of the number of countries per region. This requires taking an aggregate(average) of an aggregate (count). The subquery will first calculate the count of each country per region. The outer query will then find the average of that result.

```python
SELECT AVG(num_countries)
FROM
  (SELECT COUNT(*) AS num_countries
  FROM countries
  GROUP BY region) AS sub;
```
